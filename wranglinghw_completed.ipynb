Part #1

```python
import pandas as pd
import numpy as np
import seaborn as sns


df = pd.read_csv('./wrangling/assignment/data/airbnb_hw.csv')
df.head()



price = df['Price']
price.unique()  #A string is printed out,and commas must be removed.

price = price.str.replace(',','') # This removes all commas by replacing it with nothing
print( price.unique())
price = pd.to_numeric(price,errors='coerce') # Converts the price values, incliding strings, to a numeric type.
print( price.unique())
# # This converts all the values
print('Total missing: ', price.isna().sum())  #This checks for and prints the total number of NaN values in the price column.
```

Part #2

```python
df = pd.read_csv('/content/wrangling/data/sharks.csv', low_memory=False)
df.head()
df.columns.tolist()



type = df['Type']

type = type.replace(['Sea Disaster', 'Watercraft','Boating','Boatomg'],'Under investigation') # Because the data includes several different variations of the word "Boat", these words must be replaced wit ha single word that means the same thing. In this case I repalced them with the word "Boat".


type = type.replace(['Invalid', 'Questionable','Unconfirmed','Unverified','Pending'],np.nan) # Because these uncertain types are not useful, we can replace with NaNs using replace()
type.value_counts()

df['Type'] = type
del type         # The unclean "Type" varaiable can be replaced with the new clean "type"

df['Type'].value_counts()


df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace(['UNKNOWN', 'F','M','2017'],np.nan) # This replaces all the unclean values with NaNs
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].replace('y','Y')
pd.crosstab(df['Type'],df['Fatal (Y/N)'],normalize='index')       #This function creates a table 'Type' and 'Fatal (Y/N)', showing proportions, where the sum of values in each row adds up to 1. This shows the relative proportions of Fatal (Y/N) within each Type. This shows a better distrubution than the raw counts.
```

```python
df = pd.read_parquet('/content/wrangling/data/justice_data.parquet')
length = df['ImposedSentenceAllChargeInContactEvent']
type = df['SentenceTypeAllChargesAtConvictionInContactEvent']

print( length.unique()) #When seeing the unique values, ' ' is present, meaning that some of the values are missing
length = pd.to_numeric(length,errors='coerce') #First, because the output is a string so it must be coerce to numeric
length_NaN = length.isnull() #Indicates which entries in the length column are missing values by marking them True and non-missing values with False.

print(np.sum(length_NaN)) #This shows that 9053 thousand values are missing and need to be cleaned
print( pd.crosstab(length_NaN, type)) #Category 4 and 9  dismissed and unclear respectivly, abd therefore have missing values

length = length.mask( type == 4, 0) #This replaces length with 0 when the type is a category 4
length = length.mask( type == 9, np.nan) #This replaces length with 9 when the type is a category NaN

length_NaN = length.isnull()
print( pd.crosstab(length_NaN, type))   #Now all the missing values are replaced with the dummy variable
print( np.sum(length_NaN)) # Now only 274 values are missing instead of 9053
df['ImposedSentenceAllChargeInContactEvent'] = length #The new clean data can replace the old data
del length, type
```
